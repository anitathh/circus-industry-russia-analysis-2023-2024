```python
import pandas as pd

df_2023 = pd.read_excel("Результат_Объединения_2023.xlsx")
df_2024 = pd.read_excel("Результат_Объединения_2024.xlsx")

summary_2023 = {
    "Цирки (стационарные)": df_2023["Количество стационарных цирков, ед."].sum(),
    "Цирки (шапито)": df_2023["Количество цирков шапито, ед."].sum(),
    "Представления": df_2023["Количество цирковых представлений, ед."].sum(),
    "Зрители (тыс.)": df_2023["Количество зрителей на цирковых представлениях, тыс. чел."].sum(),
    "Билеты (тыс.)": df_2023["Количество проданных билетов, тыс. шт."].sum(),
    "Доход (тыс. руб.)": df_2023["Полученные средства от цирковых представлений, тыс. руб."].sum(),
    "Животные": df_2023["Количество животных, ед."].sum()
}

summary_2024 = {
    "Цирки (стационарные)": df_2024["Количество стационарных цирков, ед."].sum(),
    "Цирки (шапито)": df_2024["Количество цирков шапито, ед."].sum(),
    "Представления": df_2024["Количество цирковых представлений, ед."].sum(),
    "Зрители (тыс.)": df_2024["Количество зрителей на цирковых представлениях, тыс. чел."].sum(),
    "Билеты (тыс.)": df_2024["Количество проданных билетов, тыс. шт."].sum(),
    "Доход (тыс. руб.)": df_2024["Полученные средства от цирковых представлений, тыс. руб."].sum(),
    "Животные": df_2024["Количество животных, ед."].sum()
}

final_df = pd.DataFrame({
    "Показатель": summary_2023.keys(),
    "2023": summary_2023.values(),
    "2024": summary_2024.values()
})

final_df.to_excel("Для главного чарта.xlsx", index=False)

print(final_df)
```
